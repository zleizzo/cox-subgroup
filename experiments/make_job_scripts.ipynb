```python
import sys
import os
from math import ceil
sys.path.append(os.path.abspath("../src"))

from evaluation.run_experiment import get_job_list
```

```python
config_name = "table_3"
job_list = get_job_list(config_name)
print(len(job_list))
```

```python
num_jobs = len(job_list)
for i in range(ceil(num_jobs / 1000)):
    print(i, min(num_jobs - 1000 * i, 999))
    script_content = f"""#!/bin/bash
#SBATCH --job-name={config_name}
#SBATCH --cpus-per-task=12
#SBATCH --mem=16G
#SBATCH --time=02:00:00
#SBATCH --array=0-{min(num_jobs - 1000 * i - 1, 999)}
#SBATCH --output=/home/ml/zach/survival-subgroup-refactored/results/slurm_logs/{config_name}_%A_%a.out
#SBATCH --error=/home/ml/zach/survival-subgroup-refactored/results/slurm_logs/{config_name}_%A_%a.err

OFFSET={1000 * i}

# Print this sub-job's task ID
echo "My experiment number is " $((SLURM_ARRAY_TASK_ID + OFFSET))

conda init
conda activate survival
python run_slurm.py {config_name} $((SLURM_ARRAY_TASK_ID + OFFSET))
"""

    # Write to file
    with open(f"{config_name}_{i}.sh", "w") as f:
        f.write(script_content)
```
